```python
import pandas as pd
df = pd.read_json("Apache-Access-NoMessage2.json")
df.drop(['@timestamp'],axis=1,inplace=True)
df.rename(columns={'@version': 'version'},inplace=True)
df.set_index('datetime',inplace=True)
df.head(2)
addresses=list(map(lambda ip: "https://freegeoip.app/json/"+ip,df['ipaddress'].unique()))
```

```python
geoip=pd.read_csv('geoip2-ipv4.csv')
geoip.head(2)
len(addresses)
```

```python
import requests

headers = {
    'accept': "application/json",
    'content-type': "application/json"
    }

geoipdata=""
for url in addresses:
    response = requests.request("GET", url, headers=headers)
    print(response.text)
    tmpdata=response.text
    geoipdata+=tmpdata

print(geoipdata)
```

```python
f=open('geoipdata.json','w')
f.write(geoipdata)
f.close()
```
