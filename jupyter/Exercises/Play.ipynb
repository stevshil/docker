```python
apache=pd.read_json('Apache-Access-NoMessage.json')
apache
```

```python
ips=apache['ipaddress'].notnull()
apache[ips]['ipaddress'].count()
```

```python
apache[ips]['ipaddress'].drop_duplicates().count()
```

```python
import requests
geodata={}

for address in apache[ips]['ipaddress'].drop_duplicates():
    r=requests.get('https://freegeoip.app/json/'+address)
    geodata[address]=r.json()

geodata
```

```python
ipdf=pd.DataFrame(geodata)
```
